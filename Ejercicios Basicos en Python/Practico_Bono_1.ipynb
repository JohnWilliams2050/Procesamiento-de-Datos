#### MENOR DE DOS PARES: Escriba una función que devuelva el menor de dos números dados *si* ambos números son pares, pero devuelva el mayor si uno o ambos números son impares
    lesser_of_two_evens(2,4) --> 2
    lesser_of_two_evens(2,5) --> 5

```python
def lesser_of_two_evens(a,b):
    pass
    #Colocar aqui la respuesta
    if (a%2==0) and (b%2==0): ##se revisa si las dos entradas son pares
      ##despues devuelve la entrada que es la menor de las dos
      if a>b:
        return b
      elif a<b:
        return a
      else:
        return('son iguales')
    else: ##si uno o mas son impares, se devuelve cual es el mayor entre las dos
      if a>b:
        return a
      elif a<b:
        return b
      else:
        return('son iguales')

```

```python
# Revisión
lesser_of_two_evens(2,4)
```

```python
# Revisión
lesser_of_two_evens(2,5)
```

#### GALLETAS DE ANIMALES: Escribir una función que toma una cadena de dos palabras y devuelve True si ambas palabras comienzan con la misma letra
    animal_crackers('Ciudad Calor') --> True
    animal_crackers('Lugar Frio') --> False

```python
def animal_crackers(text):
  pass
  twoE = text.split() ## se separa la cadena en dos donde hay un espacio
  if twoE[0][0] ==twoE[1][0]: ##se revisa si la primera letra de ambas cadenas son iguales
    return True
  else:
    return False
```

#### HACE VEINTE: dados dos enteros, devuelve True si la suma de los enteros es 20 o si uno de los enteros es 20. De lo contrario, devuelve False

    makes_twenty(20,10) --> True
    makes_twenty(12,8) --> True
    makes_twenty(2,3) --> False

```python
def makes_twenty(n1,n2):
  pass
  if (n1==20) or (n2==20): ##revisa si alguna de las dos entradas son 20, si lo es, devuelve True
    return True
  elif n1 + n2 == 20: ##revisa si la suma de las dos entradas dan 20, si da 20, devuelve True, y si no entonces devuelve False
    return True
  else:
    return False
```

```python
makes_twenty(12,8)
```

#### MAYÚSCULAS: Escribir una función que haga mayúscula la primera y cuarta letra de una cadena de texto
     
    old_macdonald('macdonald') --> MacDonald
    
Note: `'macdonald'.capitalize()` returns `'Macdonald'`

```python
def old_macdonald(name):
    pass
    if len(name)<4: ##se revisa si se puede hacer la cuarta letra Mayuscula
      return ('no se puede imprimir la cuarta letra de la cadena de texto')
    else: ##si es mayor que 4, se divide la cadena en dos y capitaliza la primera letra y despues la cuarta letra
      new_name = name[0:3].capitalize()
      new_name_2 = name[3::].capitalize()
      new_name3 = new_name + new_name_2 ##se concatenan las dos cadenas
      return new_name3
```

```python
# Check
old_macdonald('macdonald')
```

#### REVERSA: Dada una oración, retornar la misma oración pero en reversa

    master_yoda('I am home') --> 'home am I'
    master_yoda('We are ready') --> 'ready are We'
    
Note: Revisar método .join()

    >>> "--".join(['a','b','c'])
    >>> 'a--b--c'

Resultado de .join():

    >>> " ".join(['Hello','world'])
    >>> "Hello world"

```python
def master_yoda(text):
    pass
    nText = text.split() ##se divide la cadena en una lista
    nnText = [] ##lista vacia
    size = -len(nText) ##se encuentra la longitud de la lista y se hace negativo para encontrar el indice inicial
    for i in range(size-1,-1): ##se hace un bucle comenzado desde size-1 que seria el indice inicial hasta -1 que es el indice del ultimo elemento
      ind = (i+2)*-1 ##se convierte los indices a valores positivos que estan en el rango de la lista inicial
      nnText.append(nText[ind]) ##se agrega los elementos a la nueva lista
    return nnText
```

#### PROBLEMA 33:

Dada una lista de enteros, retornar True si el arreglo tinene un  3  seguido por un  3 .

    has_33([1, 3, 3]) → True
    has_33([1, 3, 1, 3]) → False
    has_33([3, 1, 3]) → False

```python
def has_33(nums):
    pass
    check = False ##booleano que se va retornar
    indA = 0 ##indices para el primer elemento y para el elemento que le sigue
    indB = 1
    while(check == False): ##se mantiene en el bucle hasta que el check sea True
      a = nums[indA] ##el valor del primer elemento
      b = nums[indB] ##el valor del segundo elemento
      if a==3 and b==3: ##revisa si los dos elementos son igual a 3. Si lo es, se actualiza el check y se sale del bucle
        check = True
      else: ##si no lo es, entonces se actualiza el indice y se sale del bucle cuando el indice del segundo elemento es igual al indice del ultimo elemento
        if(indB==len(nums)-1):
          break
        indA = indA + 1
        indB = indB + 1
        check = False
    return check ##devuelve el booleano que determina si la lista tiene un 3 seguido por un 3
```

```python
# Revisión
has_33([1, 3, 3])
```

```python
# Revisión
has_33([1, 3, 1, 3])
```

```python
# Revisión
has_33([3, 1, 3])
```

#### REPLICADOR: Dada una cadena, devuelve una cadena donde por cada carácter en el original hay tres caracteres REPETIDOS
    paper_doll('Hello') --> 'HHHeeellllllooo'
    paper_doll('Mississippi') --> 'MMMiiissssssiiippppppiii'

```python
def paper_doll(text):
    pass
    text_n = '' ##texto nuevo
    for c in text: ##por cada caracter en el texto original, se multiplica por 3 y se concatena a la nueva cadena
      text_n = text_n + (c*3)
    return text_n
```

```python
# Check
paper_doll('Hello')
```

```python
# Check
paper_doll('Mississippi')
```

#### BLACKJACK: Dados tres enteros entre 1 y 11, si su suma es menor o igual a 21, devuelve su suma. Si su suma excede 21 *y* hay un once, reduce la suma total en 10. Finalmente, si la suma (incluso después del ajuste) excede 21, devuelve 'BUST'
    blackjack(5,6,7) --> 18
    blackjack(9,9,9) --> 'BUST'
    blackjack(9,9,11) --> 19

```python
def blackjack(a,b,c):
    pass
    sum = 0 ##la suma
    if a==11 or b==11 or c==11: ##revisa si alguno de las 3 entradas son igual a 11
      sum = a+b+c-10 ##si lo es, se le resta 10 al resultado
      if sum<=21:
        return sum ##si el resultado es menor a 21, se devuelve la suma
      else:
        return('BUST') ##si no, se devuelve BUST
    else:
      sum = a+b+c ##si ninguno es 11, solo se encuentra la suma entre los 3 valores
      if sum<=21:
        return sum ##si es menor de 21, devuelve la suma
      else:
        return('BUST') ##si es mayor a 21, devuelve BUST
```

```python
# Revisión
blackjack(5,6,7)
```

```python
# Revisión
blackjack(9,9,9)
```

```python
# Revisión
blackjack(9,9,11)
```
